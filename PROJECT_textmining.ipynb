```python
from bs4 import BeautifulSoup
import pandas as pd
import tabula
import nltk
PUB=[]
TITLE=[]
AUTHOR=[]
ABSTRACT=[]
KEYWORDS=[]
HIGHLIGHT=[]
for f in path:
    page = open(f,'r',encoding="utf8")
    page = page.read()
    soup = BeautifulSoup(page, 'html.parser')
    x=soup.find('div',attrs={'class':"Publication"})
    cells = x.text
    PUB.append(cells)
    titles=soup.find('span',attrs={'class':"title-text"})
    title = titles.text
    TITLE.append(title)
    author=soup.find('',attrs={'class':"wrapper truncated"})
    Author=author.text[31:]
    AUTHOR.append(Author)
    Highlights = soup.find("div",{"class":"Abstracts u-font-serif"},{'id':"abstracts"})
    Highlight=Highlights.text[11:]
    HIGHLIGHT.append(Highlight)
    abstract = soup.find('div',class_ = 'abstract author')
    abstract=abstract.text[8:]
    ABSTRACT.append(abstract)
    keyword = soup.find('div',class_ = 'keywords-section')
    Keyword= keyword.text[8:]
    KEYWORDS.append(Keyword)
pub_df=pd.DataFrame(PUB,columns=["publication"])
pub_df.insert(loc=0,column='Topic',value="physical_sciences")
pub_df.insert(loc=1,column='Sub-category',value="engineering")

pub_df["title"] = TITLE
pub_df["Authors"] = AUTHOR
pub_df["Highlights"]=HIGHLIGHT
pub_df["Abstracts"]=ABSTRACT
pub_df["Keywords"]=KEYWORDS
pub_df
```

```python
from nltk.corpus import stopwords
import nltk
from nltk.tokenize import sent_tokenize, word_tokenize
from nltk.text import Text
stop_words = nltk.corpus.stopwords.words('english')
newStopWords = [',', '.', "?",'%','(',')','â','€','¢','Ã','â€¢']
stop_words.extend(newStopWords)
#print(stop_words)
WORDS=[]
HAPAXES=[]
for i in ABSTRACT:
    word_tokens = word_tokenize(i)
    filtered_sentence = []
    for w in word_tokens:
        if w not in stop_words:
            filtered_sentence.append(w)
    fdist = nltk.FreqDist(filtered_sentence)
    Fdist = fdist.most_common(10)
    WORDS.append(Fdist)
    hap_axes=fdist.hapaxes()
    HAPAXES.append(hap_axes)
#print(word_tokens)
#print(filtered_sentence)
pub_df["Frequecy_dist"] = WORDS
pub_df["HAP_AXES"] = HAPAXES
pub_df
```
